```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt

women_degrees = pd.read_csv('percent-bachelors-degrees-women-usa.csv')
cb_dark_blue = (0/255,107/255,164/255)
cb_orange = (255/255, 128/255, 14/255)
stem_cats = ['Engineering', 'Computer Science', 'Psychology', 'Biology', 'Physical Sciences', 'Math and Statistics']

fig = plt.figure(figsize=(18, 3))

for sp in range(0,6):
    ax = fig.add_subplot(1,6,sp+1)
    ax.plot(women_degrees['Year'], women_degrees[stem_cats[sp]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[stem_cats[sp]], c=cb_orange, label='Men', linewidth=3)
    ax.spines["right"].set_visible(False)    
    ax.spines["left"].set_visible(False)
    ax.spines["top"].set_visible(False)    
    ax.spines["bottom"].set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(stem_cats[sp])
    ax.tick_params(bottom="off", top="off", left="off", right="off")
    
    if sp == 0:
        ax.text(2005, 87, 'Men')
        ax.text(2002, 8, 'Women')
    elif sp == 5:
        ax.text(2005, 62, 'Men')
        ax.text(2001, 35, 'Women')
plt.show()
```

```python
fig = plt.figure(figsize=(18,18))
stem_cats = ['Psychology', 'Biology', 'Math and Statistics', 'Physical Sciences', 'Computer Science', 'Engineering']
lib_arts_cats = ['Foreign Languages', 'English', 'Communications and Journalism', 'Art and Performance', 'Social Sciences and History']
other_cats = ['Health Professions', 'Public Administration', 'Education', 'Agriculture','Business', 'Architecture']

for i in range(6):
    ax = fig.add_subplot(6,3,i+1+i*2)
    ax.plot(women_degrees["Year"],women_degrees[stem_cats[i]],linewidth=3,c=cb_dark_blue,label="Women")
    ax.plot(women_degrees["Year"],100-women_degrees[stem_cats[i]],linewidth=3,c=cb_orange,label="Men")
    ax.set_title(stem_cats[i])
    ax.spines["right"].set_visible(False)
    ax.spines["bottom"].set_visible(False)
    ax.spines["top"].set_visible(False)
    ax.spines["left"].set_visible(False)
    ax.tick_params(top="off",left="off",right="off",bottom="off",labelbottom="off")
    ax.set_xlim(1968,2011)
    ax.set_ylim(0,100)
    ax.set_yticks([0,100])
    ax.axhline(50,c=(171/255, 171/255, 171/255),alpha=0.3)
    
    if i==0:
        ax.text(2008,83,"Women")
        ax.text(2008,14,"Men")
    elif i==5:
        ax.text(2008,88,"Men")
        ax.text(2008,6,"Women")
        ax.tick_params(labelbottom="on")
        
for i in range(5):
    ax = fig.add_subplot(6,3,i+2+i*2)
    ax.plot(women_degrees["Year"],women_degrees[lib_arts_cats[i]],linewidth=3,c=cb_dark_blue,label="Women")
    ax.plot(women_degrees["Year"],100-women_degrees[lib_arts_cats[i]],linewidth=3,c=cb_orange,label="Men")
    ax.set_title(lib_arts_cats[i])
    ax.spines["right"].set_visible(False)
    ax.spines["bottom"].set_visible(False)
    ax.spines["top"].set_visible(False)
    ax.spines["left"].set_visible(False)
    ax.tick_params(top="off",left="off",right="off",bottom="off",labelbottom="off")
    ax.set_xlim(1968,2011)
    ax.set_ylim(0,100)
    ax.set_yticks([0,100])
    ax.axhline(50,c=(171/255, 171/255, 171/255),alpha=0.3)
    
    if i==0:
        ax.text(2008,77,"Women")
        ax.text(2008,19,"Men")
    elif i==4:
        ax.text(2008,58,"Men")
        ax.text(2008,36,"Women")
        ax.tick_params(labelbottom="on")
        
for i in range(6):
    ax = fig.add_subplot(6,3,i+3+i*2)
    ax.plot(women_degrees["Year"],women_degrees[other_cats[i]],linewidth=3,c=cb_dark_blue,label="Women")
    ax.plot(women_degrees["Year"],100-women_degrees[other_cats[i]],linewidth=3,c=cb_orange,label="Men")
    ax.set_title(other_cats[i])
    ax.spines["right"].set_visible(False)
    ax.spines["bottom"].set_visible(False)
    ax.spines["top"].set_visible(False)
    ax.spines["left"].set_visible(False)
    ax.tick_params(top="off",left="off",right="off",bottom="off",labelbottom="off")
    ax.set_xlim(1968,2011)
    ax.set_ylim(0,100)
    ax.set_yticks([0,100])
    ax.axhline(50,c=(171/255, 171/255, 171/255),alpha=0.3)
    
    if i==0:
        ax.text(2008,90,"Women")
        ax.text(2008,6,"Men")
    elif i==5:
        ax.text(2008,64,"Men")
        ax.text(2008,28,"Women")
        ax.tick_params(labelbottom="on")
plt.savefig("gender_degrees.png")
plt.show()
```
